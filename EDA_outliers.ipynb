```python
# Install necessary packages and import relevant libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load the dataset into pandas DataFrame
df = pd.read_csv("/content/drive/MyDrive/Capstone_project/v2_credit_default.csv")
```

```python
b = sns.boxplot(x = 'LIMIT_BAL', data = df);
b.set_xlabel('LIMIT_BAL',fontsize=20)
b.tick_params(labelsize=15)
#Insight: balance limit granted to customers is skewed right, with only a few customers given very high balance limits
```

```python
repay_=['Repay_Sept', 'Repay_Aug', 'Repay_July', 'Repay_June', 'Repay_May', 'Repay_Apr']
fig,ax=plt.subplots(2,3,figsize=(20,10))

for cols,subplots in zip(repay_,ax.flatten()):
    sns.boxplot(x=df[cols],ax=subplots) 
#Insight: skewed right with only few customers late on their payment by more than 2 months
```

```python
repay_=['Bill_Sept', 'Bill_Aug', 'Bill_July', 'Bill_June', 'Bill_May', 'Bill_Apr']
fig,ax=plt.subplots(2,3,figsize=(20,10))

for cols,subplots in zip(repay_,ax.flatten()):
    sns.boxplot(x=df[cols],ax=subplots) 
#Insight: skewed right. 
```

```python
repay_=['Pay_Sept', 'Pay_Aug', 'Pay_July', 'Pay_June', 'Pay_May', 'Pay_Apr']
fig,ax=plt.subplots(2,3,figsize=(20,10))

for cols,subplots in zip(repay_,ax.flatten()):
    sns.boxplot(x=df[cols],ax=subplots) 
# Insight: more skewed right than Bill_. 
# More zeros than Bill_: There are customers that have a non-zero bill for a month, but they are not making a payment. 
```

```python
zero_bill_sept = df.Bill_Sept[df.Bill_Sept==0].count()
zero_pay_sept = df.Pay_Sept[df.Pay_Sept==0].count()
print('Zero bill in September:', zero_bill_sept )
print('Zero pay in September:', zero_pay_sept )
print ('ratio: ', zero_bill_sept/zero_pay_sept)
# Insight: 38% are not making any payments because they didn't use their credit card (thair bill was zero). The rest failed to make a payment
```
